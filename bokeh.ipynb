Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, a glyph is a graphical shape or marker that can be added to a plot to represent data points. Glyphs are the basic building blocks of Bokeh plots. Some examples of glyphs are circles, squares, lines, and text.

To add glyphs to a Bokeh plot, you can use the plot method of the figure object. Here's an example of how to add circles and lines to a Bokeh plot:



```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]


output_file("bokeh_glyphs.html")


p = figure(title="Bokeh glyphs example", x_axis_label='x', y_axis_label='y')


p.circle(x, y, size=10, color='navy', alpha=0.5)


p.line(x, y, line_width=2)


show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("bokeh_custom.html")
p = figure(title="Custom Bokeh plot", x_axis_label='x', y_axis_label='y')
p.xaxis.axis_label_text_color = "red"
p.yaxis.axis_label_text_color = "blue"
p.xaxis.major_label_text_font_size = "16pt"
p.yaxis.major_label_text_font_size = "16pt"
p.xaxis.axis_line_width = 3
p.yaxis.axis_line_width = 3
p.title.text = "My Custom Bokeh Plot"
p.title.text_font_size = "20pt"
p.title.align = "center"
p.title.text_color = "green"
p.legend.location = "top_left"
p.legend.label_text_font_size = "14pt"
p.legend.background_fill_alpha = 0.5
p.circle(x, y, size=10, color='navy', alpha=0.5, legend_label="Circles")
p.line(x, y, line_width=2, legend_label="Line")
show(p)
```

Q4.A Bokeh server is a way to create and deploy interactive Bokeh plots that can be updated in real time. With a Bokeh server, you can build complex, interactive visualizations that can be accessed from a web browser.

To create a Bokeh server application, you first define your Bokeh plot as you normally would, but instead of calling show() to display it, you create a bokeh.server.server.Server object and pass your plot to its curdoc() function. This makes your plot a part of the current document, which can be accessed and modified by the server application.

You can then define various input controls and callbacks that can update your plot in response to user interactions. For example, you can add sliders or dropdowns that update the data or appearance of your plot, or add widgets that trigger data queries or computations.


Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

You can embed a Bokeh plot into a web page or dashboard using Flask or Django by creating a Bokeh server application and embedding it as an iframe in your web page or dashboard.

To do this, you first need to create a Bokeh server application as described in the previous answer. Once you have created the Bokeh server application, you can embed it in a Flask or Django web page by adding an iframe element to the HTML template for your page, and setting the src attribute of the iframe to the URL of the Bokeh server application.

Here's an example of how to embed a Bokeh server application in a Flask web page:

```python
from flask import Flask, render_template
from bokeh.embed import server_document

app = Flask(__name__)

@app.route('/')
def index():
    bokeh_script = server_document('http://localhost:5006/myapp')
    return render_template('index.html', bokeh_script=bokeh_script)

if __name__ == '__main__':
    app.run(debug=True)
```
